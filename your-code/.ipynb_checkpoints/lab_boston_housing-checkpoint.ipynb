## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    
    
    return r2_score(y_true, y_predict)
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here

from sklearn.model_selection import train_test_split

target = data.medv
x = data.drop(columns='medv')

x_train, x_test, y_train, y_test = train_test_split(x, target, test_size = 0.2, shuffle=True)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths

from sklearn.ensemble import RandomForestRegressor


depth = [2,4,6,8,10]

results = {}
results_train = {}

for i in depth:
    ks_rf = RandomForestRegressor(n_estimators=10, max_depth = i).fit(x_train, y_train)
    y_predict = ks_rf.predict(x_test)
    y_predict_train = ks_rf.predict(x_train) 
    r2 = performance_metric(y_test, y_predict)
    r2_train = performance_metric(y_train, y_predict_train)
    results.update( {i : r2} )
    results_train.update( {i : r2_train})

```

```python
results_df = pd.DataFrame.from_dict(results, orient='index')

results_df = results_df.reset_index()

results_df_tr = pd.DataFrame.from_dict(results_train, orient='index')

results_df_tr = results_df_tr.reset_index()
```

```python
results_df = results_df.rename(columns={0:'r2_test'})

results_df
```

```python
results_df_tr = results_df_tr.rename(columns={0:'r2_train'})

results_df_tr
```

```python
results_df['r2_train'] = results_df_tr['r2_train']

results_df
```

```python
plt.plot( 'index', 'r2_test', data=results_df, marker='o', color='skyblue', linewidth=2)
plt.plot( 'index', 'r2_train', data=results_df, marker='o', color='olive', linewidth=2)
plt.legend()
```
